```python
format_type = '%H:%M:%S %d/%m/%Y'
print(data.shape)
data['f_time_from'] = pd.to_datetime(data['f_time_from'], format=format_type)
data['f_time_to'] = pd.to_datetime(data['f_time_to'], format=format_type)

data['hour'] = data['f_time_from'].dt.hour
data['day_of_week'] = data['f_time_from'].dt.day_of_week
data['day'] = data['f_time_from'].dt.day
data['month'] = data['f_time_from'].dt.month

#tinh tg bay
data['duration_minutes'] =  (data['f_time_to'] - data['f_time_from']).dt.total_seconds() / 60
data.head()
data[data['day']==30].head()
```

```python
data[(data['code']=='vj')&(data['from']=='TP HCM')&(data['to']=='Hà Nội')&(data['type']=='Eco')&(data['hour']==6)&(data['day_of_week']==6)].head(20)
```

```python

features = ['code_name', 'from', 'to', 'type']
print(data.shape)
df_encode = pd.get_dummies(data, columns=features, drop_first=True)
df_encode.shape
df_encode.columns
```

```python
print(df_encode['total_price'].value_counts())
```

```python
#!pip install catboost
from sklearn.linear_model import LinearRegression
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor


drop_cols = ['id', 'code', 'f_code', 'f_time_from', 'f_time_to',
                'f_price', 'fees', 'total_price', 'airport_from', 'airport_to']


drop_cols_time = ['id', 'code', 'f_code', 'f_time_from', 'f_time_to', 'airport_from', 'airport_to','duration_minutes']



X = df_encode.drop(columns=drop_cols)
Y = df_encode['total_price']


X_time = df_encode.drop(columns=drop_cols_time)
Y_time = df_encode['duration_minutes']


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train_time, X_test_time, Y_train_time, Y_test_time = train_test_split(X_time, Y_time, test_size=0.2, random_state=42)





print(X_train.shape)
```

# xgboost


```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error


model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1)
model.fit(X_train, Y_train)

y_predict = model.predict(X_test)
score = r2_score(Y_test, y_predict)
mae = mean_absolute_error(Y_test, y_predict)
print(f"Trung bình lệch: {mae} VND")
print(f"Do Chinh Xac(R2 score) = {score}")
```

# linear regression


```python
LR_model = LinearRegression()
LR_model.fit(X_train, Y_train)
y_predict = LR_model.predict(X_test)
score = r2_score(Y_test, y_predict)
mae = mean_absolute_error(Y_test, y_predict)
print(f"Trung bình lệch: {mae} VND")
print(f"Do Chinh Xac(R2 score) = {score}")
```

# catboost


```python
CR_model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.1,
    depth=6,
    loss_function='RMSE'
)
CR_model.fit(X_train, Y_train)
y_predict = CR_model.predict(X_test)
score = r2_score(Y_test, y_predict)
mae = mean_absolute_error(Y_test, y_predict)
print(f"Trung bình lệch: {mae} VND")
print(f"Do Chinh Xac(R2 score) = {score}")
```

# predict duration


```python
#predict duration
model_predic_time=XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1)
model_predic_time.fit(X_train_time, Y_train_time)
y_predict_time=model_predic_time.predict(X_test_time)
score = r2_score(Y_test_time, y_predict_time)
mae = mean_absolute_error(Y_test_time, y_predict_time)
print(f"Trung bình lệch: {mae} phút")
print(f"Do Chinh Xac(R2 score) = {score}")
```

```python
import ipywidgets as widgets
from IPython.display import display

# lấy danh sách lựa chọn từ dữ liệu gốc
ds_hang = data['code_name'].unique().tolist()
ds_diem_di = data['from'].unique().tolist()
ds_diem_den = data['to'].unique().tolist()
ds_loai_ve = data['type'].unique().tolist()
from datetime import datetime
# tạo widgets
w_hang = widgets.Dropdown(options=ds_hang, description='Hãng bay:')
w_di = widgets.Dropdown(options=ds_diem_di, description='Điểm đi:', value='Hà Nội')
w_den = widgets.Dropdown(options=ds_diem_den, description='Điểm đến:', value='TP HCM')
w_loai = widgets.Dropdown(options=ds_loai_ve, description='Loại vé:', value='Eco')
w_gio_di = widgets.Text(description='Giờ đi:', value=datetime.now().strftime(format_type))
w_gio_den = widgets.Text(description='Giờ đến:', value=datetime.now().strftime(format_type))
btn = widgets.Button(description="Dự đoán ngay!", button_style='success', icon='plane')
output = widgets.Output()
```

```python
def du_doan_gia_veXG(hang_bay, diem_di, diem_den, loai_ve, gio_di):
    input_df = pd.DataFrame({
        'code_name': [hang_bay],
        'from': [diem_di],
        'to': [diem_den],
        'type': [loai_ve],
        'f_time_from': [pd.to_datetime(gio_di, format=format_type)]
    })

    input_df['hour'] = input_df['f_time_from'].dt.hour
    input_df['day_of_week'] = input_df['f_time_from'].dt.day_of_week
    input_df['day'] = input_df['f_time_from'].dt.day
    input_df['month'] = input_df['f_time_from'].dt.month

    input_encoded = pd.get_dummies(input_df, columns=['code_name', 'from', 'to', 'type'])

    time_input = input_encoded.reindex(columns=X_train_time.columns, fill_value=0)
    duration_pred = float(model_predic_time.predict(time_input)[0])
    input_encoded['duration_minutes'] = duration_pred
    price_input = input_encoded.reindex(columns=X_train.columns, fill_value=0)
    predicted_price = float(model.predict(price_input)[0])

    return duration_pred,predicted_price
```

```python
# XGboost
def on_click_predict(b):
    with output:
        output.clear_output()
        try:
            time,gia = du_doan_gia_veXG(
                hang_bay=w_hang.value,
                diem_di=w_di.value,
                diem_den=w_den.value,
                loai_ve=w_loai.value,
                gio_di=w_gio_di.value
            )
            print(f"Giá vé dự đoán: {gia:,.0f} VNĐ, thời gian đi là {time:,.0f} phút")
        except Exception as e:
            print(f"Lỗi định dạng ngày tháng hoặc dữ liệu: {e}")

btn.on_click(on_click_predict)

print("--- ỨNG DỤNG DỰ ĐOÁN GIÁ VÉ ---")
display(w_hang, w_di, w_den, w_loai, w_gio_di, btn, output)
```

```python
from datetime import datetime


default_time = datetime.now().strftime(format_type)

w_hang_a = widgets.Dropdown(options=ds_hang, description='Hãng bay:')
w_di_a = widgets.Dropdown(options=ds_diem_di, description='Điểm đi:', value='Hà Nội')
w_den_a = widgets.Dropdown(options=ds_diem_den, description='Điểm đến:', value='TP HCM')
w_loai_a = widgets.Dropdown(options=ds_loai_ve, description='Loại vé:', value='Eco')
w_gio_di_a = widgets.Text(description='Giờ đi:', value=default_time, placeholder='07:00')
w_gio_den_a = widgets.Text(description='Giờ đến:', value=default_time, placeholder='09:30')
btn_a = widgets.Button(description="Dự đoán ngay!", button_style='success', icon='plane')
output_a = widgets.Output()

def du_doan_gia_veLR(hang_bay, diem_di, diem_den,loai_ve, gio_di, gio_den):
    input_df = pd.DataFrame({
        'code_name': [hang_bay],
        'from': [diem_di],
        'to': [diem_den],
        'type': [loai_ve],
        'f_time_from': [pd.to_datetime(gio_di, format=format_type)],
        'f_time_to': [pd.to_datetime(gio_den, format=format_type)],
    })

    input_df['duration_minutes'] = (input_df['f_time_to'] - input_df['f_time_from']).dt.total_seconds() / 60
    input_df['hour'] = input_df['f_time_from'].dt.hour
    input_df['day_of_week'] = input_df['f_time_from'].dt.day_of_week
    input_df['day'] = input_df['f_time_from'].dt.day
    input_df['month'] = input_df['f_time_from'].dt.month

    input_encoded = pd.get_dummies(input_df, columns=['code_name', 'from', 'to', 'type'])
    final_input = input_encoded.reindex(columns=X_train.columns, fill_value=0)

    predicted_price = model.predict(final_input)[0]
    return predicted_price


# LR
def on_click_predict_a(b):
    with output_a:
        output_a.clear_output()
        try:
            gia = du_doan_gia_veLR(
                hang_bay=w_hang_a.value,
                diem_di=w_di_a.value,
                diem_den=w_den_a.value,
                loai_ve=w_loai_a.value,
                gio_di=w_gio_di_a.value,
                gio_den=w_gio_den_a.value
            )
            print(f"Giá vé dự đoán: {gia:,.0f} VNĐ")
        except ValueError as ve:
            print(f"Lỗi: {ve}")
        except Exception as e:
            print(f"Lỗi định dạng hoặc dữ liệu: {e}")

btn_a.on_click(on_click_predict_a)

print("--- ỨNG DỤNG DỰ ĐOÁN GIÁ VÉ ---")
display(w_hang_a, w_di_a, w_den_a, w_loai_a, w_gio_di_a, w_gio_den_a, btn_a, output_a)
```
